```python
import tabula as tb
import pandas as pd
import numpy as np
import re
```

```python
file = 'Deer-Resistant-Native-Plants.pdf'
data = tb.read_pdf(file, area = (20, 0, 300, 800), pages = '2')
data = data[0]
data = data.values.tolist()
print(data)
```

```python
for index, string in enumerate(data):
    data[index] = data[index][0].split("(")
print(data)
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df.columns =['Scientific Name', 'Common Name']

df
```

```python
df["Common Name"]=df["Common Name"].str.replace(")","")
df["Scientific Name"]=df["Scientific Name"].str.replace(",","")
df
```
